```python
class Student:
    def __init__(self, name, roll_number, course, marks):
        self.name = name
        self.roll_number = roll_number
        self.course = course
        self.marks = marks

    def __str__(self):
        return f"ID: {self.id} | Name: {self.name} | Roll No: {self.roll_number} | Course: {self.course} | Marks: {self.marks}"


class StudentManager:
    def __init__(self):
        self.students = {}
        self.next_id = 1

    # CREATE
    def add_student(self, name, roll_number, course, marks):
        student = Student(name, roll_number, course, marks)
        student.id = self.next_id
        self.students[self.next_id] = student
        self.next_id += 1
        return f"Student {student.id} added successfully!"

    # READ ALL
    def view_all_students(self):
        if not self.students:
            print("No students available.")
            return
        
        for s in self.students.values():
            print(s)

    # READ ONE
    def view_student(self, student_id):
        student = self.students.get(student_id)
        if student:
            print(student)
        else:
            print(f"Student with ID {student_id} not found.")

    # UPDATE
    def update_student(self, student_id, name=None, roll_number=None, course=None, marks=None):
        student = self.students.get(student_id)

        if not student:
            print(f"Student with ID {student_id} not found.")
            return
        
        if name:
            student.name = name
        if roll_number:
            student.roll_number = roll_number
        if course:
            student.course = course
        if marks is not None:
            student.marks = marks

        print(f"Student {student_id} updated successfully!")

    # DELETE
    def delete_student(self, student_id):
        if student_id in self.students:
            del self.students[student_id]
            print(f"Student {student_id} deleted successfully!")
        else:
            print(f"Student with ID {student_id} not found.")

# TESTING SECTION


manager = StudentManager()

print(manager.add_student("Aarav", 101, "Computer Science", 89))
print(manager.add_student("Sara", 102, "IT", 92))

print("\nAll Students:")
manager.view_all_students()

print("\nUpdating Student 2:")
manager.update_student(2, marks=95)

print("\nView Student 2:")
manager.view_student(2)

print("\nAll Students:")
manager.view_all_students()

print("\nDeleting Student 1:")
manager.delete_student(1)

print("\nFinal List:")
manager.view_all_students()
```
